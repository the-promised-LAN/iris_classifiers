# Dense Neural Network Classifier for Iris Flower Dataset

## (using Keras with TensorFlow backend)

### Data loading, and pre-procesing
1 - Import data
2 - One-hot encode the flower classes
3 - Create training and testing sets

```python
import pandas as pd
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
features = ['sepal-lenght','sepal-width','petal-length','petal-width','type']
rough = pd.read_csv("data/iris.data",names=features )
```

```python
print(rough.head())
```

```python
# make a copy of the data

data = rough.copy(deep=True)
```

```python
# Create one-hot encoding for the flower types

one_hot = pd.get_dummies(data['type'])
```

```python
# Bring datasets together

data.drop('type', axis=1)
data = data.join(one_hot)
data = data.drop('type', axis=1)
```


### Split the pre-processed data into training and testing sets
- Use shuffle and split into 80% training and 20% testing


```python
X = data[features[0:4]].values
```

```python
Y = data[list(data.columns.values)[4:7]].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

## Create a dense neural net

### Define the model structure
Five fully connected layers:
1. 4 nodes (input)
2. 10 nodes
3. 15 nodes
4. 5 nodes
5. 3 nodes (output)

Last layer activated with a *softmax* since it is a multiclass classification problem

```python
model = Sequential()
model.add(Dense(10, input_dim=4, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

Ideally, would use Sigmoidal Cross-Entropy as the loss function for multiclass classification problems

```python
model.compile(loss="categorical_crossentropy", optimizer='adam', metrics=['accuracy'])
```

### Fit/train the model
Using batch size of 10 entries and 200 epochs

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=200, batch_size=10)
```
